```python
covid_df.drop(columns=['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],inplace=True)
```

```python
covid_df.head()
```

```python
covid_df.isna().sum()
```

```python
covid_df.head(5)
```

```python
covid_df['Date']=pd.to_datetime(covid_df['Date'],format='%Y-%m-%d')
```

```python
covid_df.head()
```

```python
covid_df['Active_cases']= covid_df['Confirmed']-(covid_df['Cured']+covid_df['Deaths'])
```

```python
covid_df.drop(index=range(15086,15114),inplace=True)
```

```python
covid_df.tail()
```

```python
statewise=pd.pivot_table(covid_df,values=['Confirmed','Deaths','Cured'],index=['State/UnionTerritory'],aggfunc=max)
```

```python
statewise.head()
```

```python
statewise['Recovery Rate']=(statewise['Cured']/statewise['Confirmed'])*100
```

```python
statewise['Moratility Rate']=(statewise['Deaths']/statewise['Confirmed'])*100
```

```python
statewise=statewise.sort_values(by=['Confirmed'],ascending=False)
```

```python
statewise.head()
```

```python
statewise.style.background_gradient(cmap='cubehelix')
```

```python
top_10_cases=covid_df.groupby('State/UnionTerritory')['Active_cases'].max().sort_values(ascending=False)
```

```python
fig=plt.figure(figsize=(16,9))
```

```python
index=top_10_cases.index[:10]
```

```python
values=top_10_cases.values[:10]
```

```python
plt.bar(index,values,color=['pink','red','blue'])
plt.xticks(rotation=45)
plt.xlabel('state')
plt.show()
```

```python
#Bar grap using seaborn libarary
```

```python
#top 10 states with highest death
top_10_deaths_state=covid_df.groupby('State/UnionTerritory')['Deaths'].max().sort_values(ascending=False)
```

```python
fig=plt.figure(figsize=(18,5))
plt.title("top_10_deaths_state",size=25)
ax=sns.barplot(data=top_10_deaths_state.iloc[:10],y='Deaths',x='State/UnionTerritory',linewidht=2, edgecolor='red')
plt.show()
```

```python
index=top_10_deaths_state.index[:10]
values=top_10_deaths_state.values[:10]
plt.bar(index,values)
plt.xticks(rotation=45)
plt.show()
```

```python
vaccine=vaccine_df[vaccine_df['State']!='India']
vaccine.head()
```

```python
vaccine.rename(columns={'Total Individuals Vaccinated':'Total'},inplace=True)
vaccine.head()
```

```python
# max_vaccinated_state
max_vac=vaccine.groupby('State')['Total'].sum().sort_values(ascending=False)[:5]
max_vac
```

```python
index=max_vac.index
values=max_vac.values
plt.bar(index,values,color=('red','pink','blue','orange','green'),linewidth=2,edgecolor='black')
plt.show()
```

```python
least_vac=vaccine.groupby('State')['Total'].sum().sort_values(ascending=False)[::-1]
least_vac[:5]
```
